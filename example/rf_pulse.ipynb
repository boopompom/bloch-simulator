```python
%matplotlib inline

import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
from bloch.rf_seq import hard_pulses, sinc_pulse
from bloch.simulator import sim


# Define RF pulse parameters
gamma = 26752
duration = 3.2e-3
samples = 256
dt = duration / samples
flip_angle = np.pi / 2
time_bw = 8

# Generate RF pulse
rf = sinc_pulse(time_bw, flip_angle, duration, dt)

times = np.linspace(0, duration, samples)
freqs = np.linspace(-1/dt/2, 1/dt/2, samples)
spectrum = np.fft.fftshift(np.fft.fft(rf))

# Init magnet field
g = np.ones(rf.size) * .6
dp = np.linspace(-2, 2, 512)
mx0 = np.zeros(512)
my0 = np.zeros(512)
mz0 = np.ones(512)

freq_g = dp * .6 * gamma / ( 2 * np.pi)


# Simulate
mx, my, mz = sim(rf, g, dt, 100, 100, 0, dp, 0, mx0, my0, mz0)
mxy = mx + 1.0j*my

# Calculate the spectrum of RF pulse
zoom_factor = 8
duration_ = duration * zoom_factor
samples_ = samples * zoom_factor
rf_zoomed = np.zeros(samples*zoom_factor)
for i in np.arange(samples):
    rf_zoomed[i] = rf[i]
freqs_rf = np.linspace(-1/dt/2, 1/dt/2, samples*zoom_factor)
spectrum_rf = np.fft.fftshift(np.fft.fft(rf_zoomed))

rf_ref = sinc_pulse(time_bw, np.pi/2, duration, dt)
ratio = sum(rf)/sum(rf_ref)
spectrum_g = np.interp(freq_g, freqs_rf, abs(spectrum_rf))
spectrum_g /= np.max(np.abs(spectrum_g)) / np.sin(np.pi/2 * ratio)

plt.figure(figsize=(10,8))
plt.title("Sinc RF pulse: fft vs bloch")
plt.subplot(221)
plt.xlim(0,duration)
plt.plot(times, rf)
plt.subplot(222)
plt.xlim(min(dp), max(dp))
plt.plot(dp, np.abs(spectrum_g),dp, np.abs(mxy), 'r')
plt.legend(['rf spectrum', 'spin moment'])
plt.subplot(223)
plt.ylim(min(freq_g), max(freq_g))
plt.plot(np.abs(spectrum_rf), freqs_rf)
plt.subplot(224)
plt.plot(dp, freq_g)
plt.show()

plt.figure(figsize=(10,4))
plt.plot(dp, np.angle(spectrum_g),dp, np.angle(mxy))
plt.legend(['rf spectrum', 'spin moment'])
plt.title("Phase: RF pulse spectrum vs Spin moment")
plt.show()
```

```python
# interact with jupyter notebook for changing flip angle
from ipywidgets import interact
from bokeh.io import push_notebook
from bokeh.plotting import figure, show, output_notebook, output_server
from bokeh.layouts import gridplot


figure_magnitude = figure(title=None, 
                          width=500, height=400, 
                          x_axis_label='position (cm)', y_axis_label='magnitude', 
                          y_range=(-0.2, 1.2))
rf_spectrum = figure_magnitude.line(dp, np.abs(spectrum_g), 
                                    legend="RF Pulse Spectrum", 
                                    color="red")
spin_moment = figure_magnitude.line(dp, np.abs(mxy), 
                                    legend="Bloch Spin Moment", 
                                    color="blue")

figure_phase = figure(title=None, 
                      width=400, height=400, 
                      x_axis_label='position (cm)', y_axis_label='phase')
rf_spectrum_phase = figure_phase.line(dp, np.angle(spectrum_g), 
                                      legend="RF Pulse Spectrum", 
                                      color="red")
spin_moment_phase = figure_phase.line(dp, np.angle(mxy), 
                                      legend="Bloch Spin Moment", 
                                      color="blue")

figures = gridplot([[figure_magnitude, figure_phase]], 
                   toolbar_location=None)

rf_ref = sinc_pulse(time_bw, np.pi/2, duration, dt)


def update_demo(angle=150):
    
    rf = sinc_pulse(time_bw, angle*np.pi/180, duration, dt)
    
    # Simulate
    mx, my, mz = sim(rf, g, dt, 100, 100, 0, dp, 0, mx0, my0, mz0)
    mxy = mx + 1.0j*my

    # Calculate the spectrum of RF pulse
    zoom_factor = 8
    duration_ = duration * zoom_factor
    samples_ = samples * zoom_factor
    rf_zoomed = np.zeros(samples*zoom_factor)
    for i in np.arange(samples):
        rf_zoomed[i] = rf[i]
    freqs_rf = np.linspace(-1/dt/2, 1/dt/2, samples*zoom_factor)
    spectrum_rf = np.fft.fftshift(np.fft.fft(rf_zoomed))

    ratio = sum(rf)/sum(rf_ref)
    spectrum_g = np.interp(freq_g, freqs_rf, abs(spectrum_rf))
    spectrum_g /= np.max(np.abs(spectrum_g)) / np.sin(np.pi/2 * ratio)
    
    rf_spectrum.data_source.data['y'] = np.abs(spectrum_g)
    spin_moment.data_source.data['y'] = np.abs(mxy)

    rf_spectrum_phase.data_source.data['y'] = np.angle(spectrum_g)
    spin_moment_phase.data_source.data['y'] = np.angle(mxy)
   
    push_notebook()
```
